```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
```

```python
#import CSV files into DataFrame Bandas
df1=pd.read_csv('/content/drive/MyDrive/0001Submission/Data/inDataForModels/df_TrainOutFinal.csv')
df2=pd.read_csv('/content/drive/MyDrive/0001Submission/Data/inDataForModels/df_TestOutFinal.csv')
```

```python
d1=df1.values
d2=df2.values
x_tr=d1[:,0:8]
y_tr=d1[:,8]
x_ts=d2[:,0:8]
y_ts=d2[:,8]
```

```python
gbc_params = [{ 'min_samples_split':[1000,2500],'min_samples_leaf':[10,70]}]
model = GridSearchCV(estimator = GradientBoostingClassifier(learning_rate=0.1, n_estimators=60,max_depth=9,max_features='sqrt', subsample=0.8, random_state=10),param_grid = gbc_params, scoring='accuracy',n_jobs=-1, cv=5)
model.fit(x_tr,y_tr)
predict=model.predict(x_ts)
print(model.best_params_)
print(model.best_score_)
```

```python

print("accuracy: ",accuracy_score(y_ts,predict))
print("f1 score: ",f1_score(y_ts,predict))
print("precision: ",precision_score(y_ts,predict))
print("recall: ",recall_score(y_ts,predict))
```
