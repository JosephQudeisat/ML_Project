```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
```

```python
#import CSV files into DataFrame Bandas
df1=pd.read_csv('/content/drive/MyDrive/0001Submission/Data/inDataForModels/df_TrainOutFinal.csv')
df2=pd.read_csv('/content/drive/MyDrive/0001Submission/Data/inDataForModels/df_TestOutFinal.csv')
```

```python
d1=df1.values
d2=df2.values
x_tr=d1[:,0:8]
y_tr=d1[:,8]
x_ts=d2[:,0:8]
y_ts=d2[:,8]
```

```python








model=RandomForestClassifier(n_estimators=3,n_jobs=-1)
model.fit(x_tr,y_tr)
predict=model.predict(x_ts)
print("accuracy: ",accuracy_score(y_ts,predict))
print("f1 score: ",f1_score(y_ts,predict))
print("precision: ",precision_score(y_ts,predict))
print("recall: ",recall_score(y_ts,predict))
```

```python
model=RandomForestClassifier(n_estimators=10,n_jobs=-1)
model.fit(x_tr,y_tr)
predict=model.predict(x_ts)
print("accuracy: ",accuracy_score(y_ts,predict))
print("f1 score: ",f1_score(y_ts,predict))
print("precision: ",precision_score(y_ts,predict))
print("recall: ",recall_score(y_ts,predict))
```

```python
model=RandomForestClassifier(n_estimators=20,n_jobs=-1)
model.fit(x_tr,y_tr)
predict=model.predict(x_ts)
print("accuracy: ",accuracy_score(y_ts,predict))
print("f1 score: ",f1_score(y_ts,predict))
print("precision: ",precision_score(y_ts,predict))
print("recall: ",recall_score(y_ts,predict))
```

```python
rfc = RandomForestClassifier()
#forest_params = [{ 'n_estimators': [200 ,700],'max_features': ['auto', 'sqrt', 'log2'],'max_depth': [2, 5, 15],'criterion':["gini","entropy"]}]
forest_params = [{ 'n_estimators': [20 ,60],'max_features': ['auto', 'sqrt'],'max_depth': [2,15],'criterion':["gini","entropy"]}]
model = GridSearchCV(rfc, forest_params, cv = 5, scoring='accuracy')
model.fit(x_tr,y_tr)
print(model.best_params_)
print(model.best_score_)

```

```python
predict=model.predict(x_ts)
print("accuracy: ",accuracy_score(y_ts,predict))
print("f1 score: ",f1_score(y_ts,predict))
print("precision: ",precision_score(y_ts,predict))
print("recall: ",recall_score(y_ts,predict))
```
